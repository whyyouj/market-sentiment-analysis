```python
import pickle

# Load the pickle file
with open('LSTM_pickle.pkl', 'rb') as file:
    data = pickle.load(file)

# Now you can use loaded_data
print(data['actual'])
print(data['pred'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert to DataFrames
df_actual = pd.DataFrame(data['actual'])
df_pred = pd.DataFrame(data['pred'])

# Plot the data
plt.figure()
plt.plot(df_actual, label='Actual')
plt.plot(df_pred, label='Predicted')
plt.legend()
plt.title('Actual vs Predicted')
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.show()
```

```python
df_actual.to_csv('actual.csv', index=False)
df_pred.to_csv('pred.csv', index=False)
```
